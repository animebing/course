```python
from __future__ import print_function, division
from data_utils import load_CIFAR10

X_train, y_train, X_test, y_test = load_CIFAR10("cifar-10-batches-py")

print("X_train:", X_train.shape)
print("y_train:", y_train.shape)
print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
orie = 9
ppc = (4,4)
cpb = (3,3)
```

```python
import numpy as np
from skimage.feature import hog
from skimage import color, exposure

tmp_train = []
for i in xrange(X_train.shape[0]):
#for i in range(0,2):
    gray_img = color.rgb2gray(X_train[i])
    fd = hog(gray_img, orientations=orie, pixels_per_cell=ppc,
                    cells_per_block=cpb, visualise=False, transform_sqrt=True)
    print("before", fd.shape)
    fd = fd[np.newaxis, :]
    print("after", fd.shape)
    tmp_train.append(fd)
print("tmp_train", tmp_train.shape);
X_train_hog = np.concatenate(tmp_train, axis=0)
print("X_train_hog", X_train_hog.shape);

tmp_test = []
for i in xrange(X_test.shape[0]):
    gray_img = color.rgb2gray(X_test[i])
    fd = hog(gray_img, orientations=orie, pixels_per_cell=ppc,
                    cells_per_block=cpb, visualise=False, transform_sqrt=True)
    fd = fd[np.newaxis, :]
    tmp_test.append(fd)

X_test_hog = np.concatenate(tmp_test, axis=0)

print(X_train_hog.shape)
print(X_test_hog.shape)
```

```python
from sklearn import svm
import time

tic = time.time()
lin_clf = svm.LinearSVC() 

lin_clf.fit(X_train_hog, y_train)
toc_1 = time.time()
print("time for training: %d seconds" %(toc_1-tic))

preds = lin_clf.predict(X_test_hog)
toc_2 = time.time()
print("time for prediction: %d seconds" %(toc_2-toc_1))

acc = (preds==y_test).mean()
print("accuracy: ", acc)
```

```python
from sklearn import svm

tic = time.time()
rbf_clf = svm.SVC() 

rbf_clf.fit(X_train_hog, y_train)
toc_1 = time.time()
print("time for training: %d seconds" %(toc_1-tic))

preds = rbf_clf.predict(X_test_hog)
toc_2 = time.time()
print("time for prediction: %d seconds" %(toc_2-toc_1))

acc = (preds==y_test).mean()
print("accuracy: ", acc)
```

logistic regression

```python
from sklearn import linear_model
import time
tic = time.time()
lin_reg = linear_model.LogisticRegression(multi_class='multinomial', n_jobs=-1, solver='sag', max_iter=500)
lin_reg.fit(X_train_hog, y_train)
toc_1 = time.time()
print("time for training: %d seconds" %(toc_1-tic))

preds = lin_reg.predict(X_test_hog)
toc_2 = time.time()
print("time for prediction: %d seconds" %(toc_2-toc_1))

acc = (preds==y_test).mean()
print("accuracy: ", acc)
```

```python
from sklearn import neighbors
import time 
import numpy as np

group_K = [1, 3, 5, 20, 50, 100]
acc = np.zeros((len(group_K),))
for i in xrange(len(group_K)):
    print("K: %d" %(group_K[i]))
    clf = neighbors.KNeighborsClassifier(group_K[i], n_jobs=-1)
    clf.fit(X_train_hog, y_train)
    tic = time.time()
    preds = clf.predict(X_test_hog)
    toc = time.time()
    print("time for prediction: %d seconds" %(toc-tic))

    acc[i] = (preds==y_test).mean()
    print("accuracy: ", acc[i])

print(acc)
```
