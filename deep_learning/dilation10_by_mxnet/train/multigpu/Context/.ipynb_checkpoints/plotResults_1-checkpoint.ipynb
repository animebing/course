```python
import random
import numpy as np
import os
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (8.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%reload_ext autoreload
```

```python
metrics = ['ClassAccuracy', 'ClassCrossEntropy']

def parse_log(file_name, train):
    if train:
        # train metric
        flag = 7
    else:
        # validation metric
        flag = 4
        
    train_metric = {}
    for metric in metrics:
        train_metric[metric] = []

    with open(file_name) as fin:     
        lines = fin.readlines()
        for line in lines[1:]:
            #line = line.replace('\t', ' ')
            #line = line.strip().split(' ')
            #print("line: ", line)
            line = line.strip('\n').split(' ')
            acc = line[-1].strip().split('=')
            if len(line) == flag:
                for metric in metrics:
                    if metric in acc[0]:
                          train_metric[metric].append(float(acc[-1]))
            
    return train_metric
```

```python
file_name ="model/model_1e-4/context.log"

train_acc = parse_log(file_name, True)


for idx, metric in enumerate(metrics):    
    plt.subplot(len(metrics)+1, 1, idx+1)
    plt.plot(train_acc[metric])
    plt.ylabel(metric)
    plt.xlabel('training')
    plt.grid('on')
    
plt.show()

val_acc = parse_log(file_name, False)
for idx, metric in enumerate(metrics):    
    plt.subplot(len(metrics)+1, 1, idx+1)
    plt.plot(val_acc[metric])
    plt.ylabel(metric)
    plt.xlabel('validation')
    plt.grid('on')
    
plt.show()
```
